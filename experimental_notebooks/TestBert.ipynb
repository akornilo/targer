```python
from dataclasses import dataclass
from typing import List, Optional, Union
@dataclass
class InputFeatures:
    """
    A single set of features of data.
    Property names are the same names as the corresponding inputs to a model.
    """

    input_ids: List[int]
    attention_mask: List[int]
    token_type_ids: Optional[List[int]] = None
    label_ids: Optional[List[int]] = None
```

```python
from torch.utils.data.dataset import Dataset

class TokenClassificationDataset(Dataset):
    
    def __init__(self, sentences, labels):
        all_data = []
        final_labels = []
        for sentence, sent_labels in zip(sentences, labels):
            try:
                tokenized_text = tokenizer(sentence , padding=True, return_attention_mask=True, is_split_into_words=True)

                final_labels = [('<pad>', '', '[CLS]', 101)]

                cur_idx = -1

                for subword_id in tokenized_text['input_ids'][1:-1]:

                    subword = tokenizer.ids_to_tokens[subword_id]        

                    if subword.startswith('##'):
                        final_labels.append((sent_labels[cur_idx], sentence[cur_idx], subword, subword_id))
                    else:
                        cur_idx += 1
                        final_labels.append((sent_labels[cur_idx], sentence[cur_idx], subword, subword_id))  

                final_labels.append(('<pad>', '', '[SEP]', 102))
                all_data.append(final_labels)  
            except IndexError:
                print("bad sample")
                continue # Just skip the annoying examples
                
        print("total examples", len(all_data))
        rearranged_inp = []
        labels = []
        for example in all_data:
            my = [i[3] for i in example]
            rearranged_inp.append(torch.Tensor(my).long())
            myl = [class_map[i[0]] for i in  example]
            labels.append(torch.Tensor(myl).long())
        
        X = pad_sequence(rearranged_inp, batch_first=True)
        y = pad_sequence(labels, batch_first=True)
        
        self.features = []
        for i in range(len(y)):
            feat = InputFeatures(input_ids=X[i], attention_mask=X[i] == 0, label_ids=y[i])
            self.features.append(feat)
            
    def __len__(self):
        return len(self.features)

    def __getitem__(self, i):
        return self.features[i]
```
