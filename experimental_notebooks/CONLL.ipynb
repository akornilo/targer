```python
import os
ALL_CONLL = {}
prefix = '../external/targer/data/AM/web_discourse/gold.data.toulmin/'
for file in os.listdir(prefix):
    data = BeautifulSoup(open(prefix + file), 'lxml')
    final_list = []
    text = data.find('cas:sofa').attrs['sofastring']

    premises = []
    for d in data.find_all('types:premise'):
        premises.append((int(d['begin']), int(d['end'])))

    claims = []    
    for d in data.find_all('types:claim'):
        claims.append((int(d['begin']), int(d['end'])))

    idx = 1
    in_claim = False
    in_premise = False
    for token in data.find_all('type4:token'):
        i = int(token.attrs['begin'])
        j = int(token.attrs['end'])

        marker = 'O'
        for start, end in premises:
            if start == i:
                marker = 'B-Premise'
                if j != end:
                    in_premise = True

            elif end == 'j' and in_premise:
                marker = 'I-Premise'
                in_premise = False
            elif i > start and j < end:
                marker = 'I-Premise'

        for start, end in claims:
            if start == i:
                marker = 'B-Claim'
                if j != end:
                    in_claim = True
            elif end == 'j' and in_claim:
                marker = 'I-Claim'
                in_claim = False
            elif i > start and j < end:
                marker = 'I-Claim'
        final_list.append([idx, text[i:j], marker])

        idx += 1
    ALL_CONLL[file] = final_list
```

```python
with open('test.conll', 'w') as f:
    for lines in ALL_CONLL.values():
        for line in lines:
            f.write(f'{line[0]}\t{line[1]}\t{line[2]}')
            f.write('\n')
        f.write('\n')
```
